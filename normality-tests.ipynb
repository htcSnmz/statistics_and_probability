```python
np.random.seed(10)
data = stats.norm.rvs(loc=38, size=500)
data[0:20]
```

```python
# by kurtosis and skewness (between 1.5 and -1.5 => data is normal)
print(stats.kurtosis(data))
print(stats.skew(data))
```

```python
# with histogram chart
sns.distplot(data)
plt.show()
```

```python
# with qqplot
stats.probplot(data, dist="norm", plot=plt)
plt.show()
```

```python
# kolmogorov-smirnov test
ks = stats.kstest(data, cdf="norm", args=(data.mean(), data.std()))
print(ks) # p_value > 0.05 => data is normal
print(f"{ks.pvalue:5f}")
```

```python
# shapiro-wilk test
sw = stats.shapiro(data)
print(sw) # p_value > 0.05 => data is normal
print(f"{sw.pvalue:5f}")
```
